```python
# Your code here

# Complex example

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
data = pd.read_csv("exploration_data/salary_data.csv")

# Extract variables
experience = data["YearsExperience"]
salary = data["Salary"]

# Prepare data for regression
X = experience.values.reshape(-1, 1)
y = salary.values
model = LinearRegression()
model.fit(X, y)
predicted_salary = model.predict(X)

# Define figure and subplots
plt.style.use('seaborn-v0_8-darkgrid')
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Comprehensive Visualization of Salary vs Experience", fontsize=16, fontweight='bold')

# Subplot 1: Basic scatter plot
axs[0, 0].scatter(experience, salary, color='blue', edgecolor='black', alpha=0.7, label='Actual')
axs[0, 0].plot(experience, predicted_salary, color='red', linewidth=2.5, label='Regression Line')
axs[0, 0].set_title("Scatter + Regression Line")
axs[0, 0].set_xlabel("Years of Experience")
axs[0, 0].set_ylabel("Annual Salary")
axs[0, 0].legend()
axs[0, 0].grid(True)

# Subplot 2: Colored scatter by quantile
colors = pd.qcut(salary, q=4, labels=['Low', 'Med-Low', 'Med-High', 'High'])
color_map = {'Low': 'skyblue', 'Med-Low': 'green', 'Med-High': 'orange', 'High': 'red'}
axs[0, 1].scatter(experience, salary, c=colors.map(color_map), s=100, alpha=0.6, edgecolor='black')
axs[0, 1].set_title("Salary Colored by Quartile")
axs[0, 1].set_xlabel("Years of Experience")
axs[0, 1].set_ylabel("Annual Salary")

# Add annotation
highlight_idx = salary.idxmax()
axs[0, 1].annotate("Highest Earner", xy=(experience[highlight_idx], salary[highlight_idx]),
                   xytext=(experience[highlight_idx]+1, salary[highlight_idx]+5000),
                   arrowprops=dict(facecolor='black', arrowstyle="->"))

# Subplot 3: Histogram of salary
axs[1, 0].hist(salary, bins=10, color='slateblue', edgecolor='white')
axs[1, 0].set_title("Histogram of Salaries")
axs[1, 0].set_xlabel("Salary")
axs[1, 0].set_ylabel("Frequency")

# Subplot 4: Boxplot
axs[1, 1].boxplot([salary, experience], labels=["Salary", "Experience"], patch_artist=True,
                  boxprops=dict(facecolor='lightgreen', color='darkgreen'),
                  medianprops=dict(color='red'),
                  whiskerprops=dict(color='gray'))
axs[1, 1].set_title("Boxplot of Salary and Experience")

# Formatting all y-axes to include thousands separator
for ax in axs.flatten():
    ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f"${x:,.0f}"))

# Adjust layout and save
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("salary_experience_analysis.png", dpi=300)
plt.show()
```
