### Example of multiple visualizations in one

```python
# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Histogram of Player Ratings
axs[0, 0].hist(df['Rating'], bins=10, edgecolor='black')
axs[0, 0].set_title('Distribution of Player Ratings')
axs[0, 0].set_xlabel('Rating')
axs[0, 0].set_ylabel('Frequency')

# Plot 2: Bar Plot of Nationalities
df['Nationality'].value_counts().head(10).plot(kind='bar', ax=axs[0, 1])
axs[0, 1].set_title('Number of Players by Nationality (top 10)')
axs[0, 1].set_xlabel('Nationality')
axs[0, 1].set_ylabel('Number of Players')

# Plot 3: Scatter Plot of Height vs. Weight
axs[1, 0].scatter(df['Height'], df['Weight'], color='blue')
axs[1, 0].set_title('Height vs Weight')
axs[1, 0].set_xlabel('Height (cm)')
axs[1, 0].set_ylabel('Weight (kg)')

# Plot 4: Box Plot of Player Ages
df['Age'].plot(kind='box', ax=axs[1, 1])
axs[1, 1].set_title('Age Distribution of Players')
axs[1, 1].set_ylabel('Age')


# Adjust layout to prevent overlapping
plt.tight_layout()

# Display the plot
plt.show()
```

## Part 2 - Lab
Now that you know how to use Matplotlib, it is time to get back to helping the marketing manager. In the folder "data" you can find the .csv file we generated yesterday. As mentioned before, the marketing manager is interested in knowing which sports have gotten most publicity in the past few years and hence don't need more publicity and which ones have not gotten enough publicity. There is two visualisation we would like you to produce:

1. a bar chart that allows the marketing manager to compare the mentions of the featured sports
2. a line chart that tracks the change in mentions of the top 3 mentioned sports over the past three years

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
df = pd.read_csv('lab_data/article_previews.csv')

# Display the first few rows to understand the structure
print(df.head())
```

<div style='background: #6A9EB4; padding: 10px; border: thin solid darblue; border-radius: 5px'>


We need to prepare the data for visualization:

1. Count the total mentions of each sport (tag) over all the years.
2. Identify the top 3 mentioned sports.
3. Track the change in mentions of these top 3 sports over the past three years.

```python
df['Tag'].value_counts()
```

```python
# get total mentions
mentions_by_sport = df['Tag'].value_counts().head(15) # limit to 15 for readability


plt.figure(figsize=(12, 6))

mentions_by_sport.plot(kind='bar', color='skyblue')

plt.title('Total Mentions of Each Sport')
plt.xlabel('Sport')
plt.ylabel('Number of Mentions')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df['Year'] = df['Year'].astype(str)


top_3_sports = mentions_by_sport.head(3).index

# filtering
df_top_3 = df[df['Tag'].isin(top_3_sports)]

# Group by year and sport, then count mentions
mentions_by_year = df_top_3.groupby(['Year', 'Tag']).size().unstack().fillna(0)

# Create a line chart
plt.figure(figsize=(12, 6))
for sport in mentions_by_year.columns:
    plt.plot(mentions_by_year.index, mentions_by_year[sport], marker='o', label=sport)

plt.title('Mentions of Top 3 Sports Over the Past Three Years')
plt.xlabel('Year')
plt.ylabel('Number of Mentions')
plt.xticks(rotation=45)
plt.legend(title='Sport')
plt.tight_layout()
plt.show()
```
