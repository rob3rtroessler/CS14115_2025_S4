# Introduction to Linear and Logistic Regression in Python

    
## Why Use Linear and Logistic Regression?

Regression analysis is a powerful statistical technique that allows us to understand and predict relationships between variables. Linear and logistic regressions are two fundamental types of regression analysis widely used in various fields, such as finance, healthcare, and social sciences. Linear regression is used for predicting continuous outcomes, while logistic regression is used for predicting categorical outcomes.



<div style='background: lightsalmon; padding: 10px; border: thin solid darblue; border-radius: 5px'>

## Linear Regression

### Theory and Mathematics

Linear regression aims to model the relationship between a dependent variable \(y\) and one or more independent variables \(X\) by fitting a linear equation to observed data. The equation of a simple linear regression line is:

$$y = \beta_0 + \beta_1 X + \epsilon$$

where:
- $y$ is the dependent variable.
- $X$ is the independent variable.
- $\beta_0$ is the intercept.
- $\beta_1$ is the slope of the line.
- $\epsilon$ is the error term.

### Use Cases and Benefits

**Use Cases:**
- Predicting housing prices based on various features.
- Estimating stock prices based on historical data.
- Forecasting sales for a company.

**Benefits:**
- Simple and easy to interpret.
- Computationally efficient.
- Provides insights into the relationship between variables.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(0)
days = np.arange(1, 21)
prices = 100 + days * 2 + np.random.normal(0, 2, days.shape[0])

# Create a DataFrame
data = pd.DataFrame({'Day': days, 'Price': prices})

# Plot the data
plt.scatter(data['Day'], data['Price'], color='blue')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Stock Prices Over Time')
plt.show()

# Perform linear regression
X = data[['Day']]
y = data['Price']
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

# Plot the regression line
plt.scatter(data['Day'], data['Price'], color='blue')
plt.plot(data['Day'], predictions, color='red')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Stock Prices Over Time with Linear Regression Line')
plt.show()
```

## Logistic Regression

### Theory and Mathematics

Logistic regression is used when the dependent variable is categorical. The goal is to model the probability that a given observation belongs to a particular category. The logistic function (sigmoid function) is used to map predicted values to probabilities:

$$P(y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X)}}$$

### Use Cases and Benefits

**Use Cases:**
- Classifying emails as spam or not spam.
- Predicting whether a student will pass or fail based on study hours.
- Determining if a patient has a disease based on diagnostic features.

**Benefits:**
- Suitable for binary and multiclass classification problems.
- Provides probabilities for class membership.
- Can handle non-linear relationships with appropriate transformations.

### Example: Logistic Regression with Synthetic Olympics Data

Let's create a synthetic dataset for the Olympics and perform logistic regression analysis.

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# generate synthetic data
np.random.seed(0)
n_athletes = 20
hours_trained = np.random.randint(1, 10, size=n_athletes)
medal_won = (hours_trained > 5).astype(int)

# create a DataFrame
data = pd.DataFrame({'HoursTrained': hours_trained, 'MedalWon': medal_won})

# plot the data
plt.scatter(data['HoursTrained'], data['MedalWon'], color='blue')
plt.xlabel('Hours Trained')
plt.ylabel('Medal Won (1=Yes, 0=No)')
plt.title('Olympics Medal Winning Prediction')
plt.show()

# perform logistic regression
X = data[['HoursTrained']]
y = data['MedalWon']
model = LogisticRegression()
model.fit(X, y)

# generate a finer grid of values for HoursTrained
hours_trained_fine = np.linspace(1, 9, 100).reshape(-1, 1)
predictions_fine = model.predict_proba(hours_trained_fine)[:, 1]

# plot the logistic regression curve
plt.scatter(data['HoursTrained'], data['MedalWon'], color='blue')
plt.plot(hours_trained_fine, predictions_fine, color='red')
plt.xlabel('Hours Trained')
plt.ylabel('Medal Won Probability')
plt.title('Olympics Medal Winning Prediction with Logistic Regression')
plt.show()
```

## Comparison of Linear and Logistic Regression

Linear and logistic regressions serve different purposes and are used in different contexts:

- **Linear Regression:**
  - Predicts continuous outcomes.
  - Best for relationships that are linear.
  - Simple interpretation and implementation.

- **Logistic Regression:**
  - Predicts categorical outcomes (binary or multiclass).
  - Models probabilities and provides a probabilistic framework.
  - Can handle non-linear relationships with transformations.

Both methods are foundational in data analysis and machine learning, providing essential tools for understanding and predicting various types of data.
