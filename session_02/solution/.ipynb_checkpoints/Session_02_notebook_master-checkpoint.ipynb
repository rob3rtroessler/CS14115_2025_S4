```python
# find the title
article_previews[0].find("h1").text
```

```python
# find numbers of authors
authors_soup = article_previews[0].find_all("a", class_="css-1fvj8jv")
number_of_authors = len(authors_soup)
print(number_of_authors)
```

```python
# names of authors in a list
authors_list = []
for author_soup in authors_soup:
    authors_list.append(author_soup.text)
print(authors_list)
```

```python
# tag
tag = article_previews[0].find("div", class_="subsection css-mtdwtq").text
print(tag)
```

```python
# Assume article_previews is a list of HTML elements containing preview data
# We are extracting the href attribute from the anchor tag within the third element
date_string_unprocessed = str(article_previews[0].find('a').get('href'))
print(date_string_unprocessed)  # Print the raw URL to verify the extraction

# Define the regex pattern for extracting the date in the URL
#   (\d{4}): Matches exactly four digits (the year) and captures them in a group
#   /: Matches the literal "/" character
#   (\d{1,2}): Matches one or two digits (the month) and captures them in a group
#   /: Matches the literal "/" character
#   (\d{1,2}): Matches one or two digits (the day) and captures them in a group
#   /: Matches the literal "/" character
pattern = r"/(\d{4})/(\d{1,2})/(\d{1,2})/"

# Use re.search to find the pattern in the URL and extract the groups
# re.search returns a Match object if the pattern is found
# .groups() method returns a tuple containing all the captured groups
year, month, day = re.search(pattern, date_string_unprocessed).groups()

# Print the extracted year, month, and day to verify the extraction
print(year)
print(month)
print(day)
```

## Task #2 - generate a file with information for the marketing manager

OK, now that you know how to retrieve this information, write a program that extracts this information and writes it into the csv file "article_previews.csv". If you can't quite remember how you create and write into files, take a look at yesterday's assignment.

```python
file_write = open('article_previews.csv', 'w', newline='')

writer = csv.writer(file_write)

for preview in article_previews:

    # title
    title = preview.find("h1").text
    
    # tag
    tag = preview.find("div", class_="subsection css-mtdwtq").text
    
    # author info    
    authors_soup = preview.find_all("a", class_="css-1fvj8jv")
    number_of_authors = len(authors_soup)
    authors_list = []
    for author_soup in authors_soup:
        authors_list.append(author_soup.text)
   
    # year, month, day
    date_string_unprocessed = str(preview.find('a').get('href'))
    #print(title, tag, date_string_unprocessed)
    pattern = r"/(\d{4})/(\d{1,2})/(\d{1,2})/"
    year, month, day = re.search(pattern, date_string_unprocessed).groups()
    
    preview_line = [title, tag, number_of_authors, authors_list,year, month, day]
    writer.writerow(preview_line)

# close the files
file_write.close()
```

## Bonus Tasks
Congratulations! You have created a csv file, which we can use to explore our extracted data. So far, the file, however, only features the most recent articles, i.e. the articles from page 1 of the Sport's Features section. Write a program that creates a csv file with tag, title, and year of all articles from 2021-2019.


```python
article_previews=[]
for x in range(1, 12):
    page = requests.get("https://www.thecrimson.com/tag/sports-features/page/{0}/".format(x))
    print(page.url)
    soup = BeautifulSoup(page.content, "html.parser")
    article_list = soup.find_all("div", class_="css-1w04wz2")
    for element in article_list:
        article_previews.append(element)
    
print(len(article_previews))
```

```python
file_write = open('article_previews.csv', 'w', newline='')

writer = csv.writer(file_write)

for preview in article_previews:

    # title
    title = preview.find("h1").text
    
    # tag
    tag = preview.find("div", class_="subsection css-mtdwtq").text
    
    # author info    
    authors_soup = preview.find_all("a", class_="css-1fvj8jv")
    number_of_authors = len(authors_soup)
    authors_list = []
    for author_soup in authors_soup:
        authors_list.append(author_soup.text)
   
    # year, month, day
    date_string_unprocessed = str(preview.find('a').get('href'))
    #print(title, tag, date_string_unprocessed)
    pattern = r"/(\d{4})/(\d{1,2})/(\d{1,2})/"
    year, month, day = re.search(pattern, date_string_unprocessed).groups()
    
    preview_line = [title, tag, number_of_authors, authors_list,year, month, day]
    writer.writerow(preview_line)

# close the files
file_write.close()
```
